```python
movies.columns = ['Film','Genre','RottenTomatoesRatings','AudienceRatings','Budget','YearOfRelease']
movies
```

```python
#Above, YearOfRelease is given. This YearOfRelease must be of type category. Otherwise, It doesn't make any sense..

#Convert YearOfRelease into category..

movies.YearOfRelease = movies.YearOfRelease.astype('category') 
movies['Film'] = movies['Film'].astype('category')
movies['Genre'] = movies['Genre'].astype('category')
```

```python
#Another example...

plt.hist([movies[movies.Genre=='Drama'].AudienceRatings,\
          movies[movies.Genre=='Action'].AudienceRatings, \
          movies[movies.Genre=='Comedy'].AudienceRatings],\
          bins=10, stacked=True)
plt.show()
```

```python
#When stacked not equal to TRUE
plt.hist([movies[movies.Genre=='Drama'].AudienceRatings,\
          movies[movies.Genre=='Action'].AudienceRatings, \
          movies[movies.Genre=='Comedy'].AudienceRatings],\
          bins=10)
plt.show()
```

```python
#Same using function...
#Using Charts 4..

list1 = []
mylabel = [] 

for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==gen].AudienceRatings)
    mylabel.append(gen)

plt.hist(list1, bins=10, stacked=True,label=mylabel, rwidth=1)
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style('darkgrid')
f, axes = plt.subplots(2,2, figsize=(15,15))

k1 = sns.kdeplot(movies.Budget, movies.AudienceRatings, shade=True, shade_lowest=False, cmap='Reds',ax=axes[0,0])
k2 = sns.kdeplot(movies.Budget, movies.RottenTomatoesRatings, shade=True, shade_lowest=False, cmap='Reds', ax=axes[0,1])
axes[1,0].hist(movies[movies.Genre=='Action'].AudienceRatings, bins=10)
k1.set(ylim=(-20,120))
k3 = sns.violinplot(data=movies, x='Genre', y='AudienceRatings', ax=axes[1,1])
```
